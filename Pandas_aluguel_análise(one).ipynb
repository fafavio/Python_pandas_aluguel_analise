Criando colunas numéricas

```python
url = 'https://raw.githubusercontent.com/alura-cursos/pandas-conhecendo-a-biblioteca/main/base-de-dados/aluguel.csv'
dados = pd.read_csv(url, sep=';')
```

```python
dados.head()
```

```python
#da para somar duas colunas para extrair, por exemplo, o valor por mês.
dados['Valor_por_mes'] = dados['Valor'] + dados['Condominio']


```

```python
# fazendo agora valor por ano

dados['Valor_por_ano'] = dados['Valor_por_mes'] * 12 + dados['IPTU']
dados.head()
```

Criar colunas categoricas

```python
#informações grudadas
dados['Descricao'] = dados['Tipo'] + dados['Bairro']
dados.head()
```

```python
# para separar as informações bonitinho

dados['Descricao'] = dados['Tipo'] + ' em ' + dados['Bairro']
dados.head()
```

```python
# se quiser da para por mais informações ainda

dados['Descricao'] = dados['Tipo'] + ' em ' + dados['Bairro'] + ' com ' + \
                                        dados['Quartos'] + ' quarto(s) ' + \
                                        ' e ' + dados['Vagas'] + ' vaga(s) de garagem.'
#vai dar erro pois estamos usando colunas com tipos diferentes como numéricas e strings
```

```python
#Transformando tudo para string vai funcionar corretamente

dados['Descricao'] = dados['Tipo'] + ' em ' + dados['Bairro'] + ' com ' + \
                                        dados['Quartos'].astype(str) + ' quarto(s) ' + \
                                        ' e ' + dados['Vagas'].astype(str) + ' vaga(s) de garagem.'
dados.head()
```

```python
#criando coluna Possuí suíte

dados['Possui_suite'] = dados['Suites'].apply(lambda x: "Sim" if x > 0 else "Não")
dados.head()
```

```python
#salvar a base de dados

dados.to_csv('dados_completos_dev.csv', index=False, sep=';')
```
